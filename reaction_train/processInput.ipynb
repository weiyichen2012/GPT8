```python
import pandas as pd
import numpy as np
import numpy.random as random
```

```python
emotionList = ['happy', 'sad', 'neutral']
sceneList = ['study', 'gaming', 'idle']

actionList = ['keep_still', 'pat_pat', 'dance', 'sing']
lightList = ['red', 'blue', 'white', 'yellow']
```

```python
def generatePreference(ifPreferred):
  # r = random.random()
  # if ifPreferred == 1:
  #   if r > 0.1:
  #     return 1
  #   elif r > 0.02:
  #     return 0
  #   else:
  #     return 0
  # elif ifPreferred == 0:
  #   if r > 0.9:
  #     return 1
  #   elif r > 0.1:
  #     return 0
  #   else:
  #     return -1
  # elif ifPreferred == 0:
  #   if r > 0.1:
  #     return 0
  #   elif r > 0.02:
  #     return 0
  #   else:
  #     return 1
  return ifPreferred

def checkFit(left, right):
  if left == 'any' or right == 'any':
    return True
  return left == right

def checkPattern(result, pattern_list):
  for pattern in pattern_list:
    if checkFit(result[0][0], pattern[0]) and checkFit(result[0][1], pattern[1]) and checkFit(result[1][0], pattern[2]) and checkFit(result[1][1], pattern[3]):
      return True, pattern[4]
  return False, ''

def checkPreference(emotion, scene, action, light, preference_list):
  result = 0
  for preference in preference_list:
    if checkFit(emotion, preference[0]) and checkFit(scene, preference[1]) and checkFit(action, preference[2]) and checkFit(light, preference[3]):
      result += preference[4]
  if result > 0:
    return 1
  elif result == 0:
    return 0
  else:
    return -1
  # return generatePreference(result)

def generateInput(input_size, preference_list, pattern_list):
  result = []
  for i in range(input_size):
    if i >= 2:
      ifPattern, patternEmotion = checkPattern(result[i-2:], pattern_list)
      if ifPattern:
        emotion = patternEmotion
      else:
        emotion = random.choice(emotionList)
    else:
        emotion = random.choice(emotionList)
    scene = random.choice(sceneList)
    action = random.choice(actionList)
    light = random.choice(lightList)
    prefrence = checkPreference(emotion, scene, action, light, preference_list)
    result.append([emotion, scene, action, light, prefrence])
  return result
    
```

```python
# 1 represend preference, 0 represent dislike
preference_list = [
  ['sad', 'any', 'pat_pat', 'any', 1],
  ['happy', 'any', 'pat_pat', 'any', -1],  
  ['sad', 'any', 'dance', 'any', -1],
  ['happy', 'any', 'dance', 'any', 1],  

  ['sad', 'any', 'any', 'red', -1],
  ['happy', 'any', 'any', 'red', 1],  
  ['sad', 'any', 'any', 'blue', 1],
  ['happy', 'any', 'any', 'blue', -1],

  ['any', 'gaming', 'dance', 'any', -1],
  ['any', 'idle', 'dance', 'any', 1],
  ['any', 'gaming', 'keep_still', 'any', 1],
  ['any', 'idle', 'keep_still', 'any', -1],
]

pattern_list = [
  ['any', 'idle', 'any', 'idle', 'sad'],
  ['happy', 'any', 'happy', 'any', 'neutral'],
  ['sad', 'any', 'sad', 'any', 'happy'],
]
```

```python
result = generateInput(1000, preference_list, pattern_list)
df = pd.DataFrame(result, columns=['emotion', 'scene', 'action', 'light', 'preference'])
df.head(10)
```
